```python
from Network import Network, Non_Linearity_Functions, Optimizers
import json

with open('../data/mnist_handwritten_train.json', 'r') as f:
    data = json.loads(f.read())

print(len(data[0]['image']) == 28*28)
print(data[0].keys())
```

```python
import numpy as np

train = []

for pair in data:
    train += [(np.array(pair['image'])/255, np.array([(1 if i == pair['label'] else 0) for i in range(10)]))]
    

with open('../data/mnist_handwritten_test.json', 'r') as f:
    data = json.loads(f.read())

test = []

for pair in data:
    test += [(np.array(pair['image'])/255, np.array([(1 if i == pair['label'] else 0) for i in range(10)]))]
```

```python
net = Network([28*28, 30, 30, 10, 10])

net.SGD_train(train, 30, 2, 50, test=test, optimizer=Optimizers.constant_eta)
```

```python
net = Network([28*28, 30, 30, 10, 10])
net.SGD_train(train, 30, 2, 50, test=test, optimizer=Optimizers.nag)
```

```python
net = Network([28*28, 30, 30, 10, 10])
net.SGD_train(train, 30, 2, 50, test=test, optimizer=Optimizers.momentum)
```

```python
net = Network([28*28, 30, 10])
net.SGD_train(train, 30, 2, 50, test=test, optimizer=Optimizers.constant_eta)
```
